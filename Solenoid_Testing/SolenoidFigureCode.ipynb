```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'/Users/rajharsora/Documents/DataImportLeeSolenoidTesting.csv')
df_time = df.groupby('Time').mean().reset_index()
```

```python
df_time = df_time.sort_values('ul', ascending=True)
```

```python
print('Dataset: \n', df.head())
print('\n Averages: \n', df_time.head())
```

```python
df_std = df.groupby('Time').std().reset_index()
print('\n Standard Deviations: \n', df_std.head())
```

```python
order = df_time['Time']
fig, ax = plt.subplots(1, figsize=(12,10))
ax1 = sns.scatterplot(y="ul", x="Time", data=df_time, marker='_', s=1000, color='blue')
ax2 = sns.swarmplot(data = df, x = 'Time', y = 'ul', color = 'k', order = order, zorder=0, ax=ax1)
```

```python
from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], marker='o', color='w', 
                          label='Test result', 
                          markerfacecolor='k', markersize=10),
                   Line2D([0], [0], marker='_', color='blue', 
                          label='Mean Water Dispensed', 
                          linestyle='None', markersize=25)]
legend = ax.legend(handles=legend_elements, loc='upper right', 
                       ncol=2, frameon=False, fontsize=12)
```

```python
ax.set_title("Lee Solenoid Testing Results")
```
